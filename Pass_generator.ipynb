```python
import random
import string
import tkinter as tk
from tkinter import messagebox

class PasswordGeneratorApp:
    def __init__(self, root):
        self.root = root
        self.root.title("Password Generator App")

        # Frame for password length and complexity
        options_frame = tk.Frame(root)
        options_frame.pack(pady=10)

        # Labels for password length and complexity
        self.length_label = tk.Label(options_frame, text="Password Length:")
        self.length_label.grid(row=0, column=0, padx=10)

        self.complexity_label = tk.Label(options_frame, text="Password Complexity:")
        self.complexity_label.grid(row=0, column=1, padx=10)

        # Text entry for password length
        self.length_entry = tk.Entry(options_frame)
        self.length_entry.grid(row=1, column=0, padx=10)

        # Dropdown menu for password complexity
        self.complexity_var = tk.StringVar()
        self.complexity_var.set("Medium")
        self.complexity_menu = tk.OptionMenu(options_frame, self.complexity_var, "Low", "Medium", "High")
        self.complexity_menu.grid(row=1, column=1, padx=10)

        # Button to generate password
        self.generate_button = tk.Button(root, text="Generate Password", command=self.generate_password)
        self.generate_button.pack(pady=20)

        # Label to display generated password
        self.password_label = tk.Label(root, text="", font=("Helvetica", 16, "bold"))
        self.password_label.pack()

    def generate_password(self):
        # Get password length from entry
        length = self.length_entry.get()

        if not length.isdigit():
            messagebox.showerror("Error", "Please enter a valid password length.")
            return

        length = int(length)
        complexity = self.complexity_var.get()

        # Define character sets based on complexity
        if complexity == "Low":
            chars = string.ascii_letters
        elif complexity == "Medium":
            chars = string.ascii_letters + string.digits
        else:
            chars = string.ascii_letters + string.digits + string.punctuation

        # Generate password
        password = "".join(random.choice(chars) for _ in range(length))
        self.password_label.config(text="Generated Password: " + password)

if __name__ == "__main__":
    root = tk.Tk()
    app = PasswordGeneratorApp(root)
    root.mainloop()
```
